```python
import numpy as np
import matplotlib.pyplot as plt
import cv2, warnings

warnings.filterwarnings("ignore")
```

```python
original_img = cv2.imread(r"./Dataset\I-HAZE\I-HAZY NTIRE 2018\GT/02_indoor_GT.jpg")

plt.imshow(original_img)
plt.title("Original-image")
plt.show()
```

```python
hazy_img = cv2.imread(r".\Dataset\I-HAZE\I-HAZY NTIRE 2018\hazy/02_indoor_hazy.jpg")
plt.imshow(hazy_img)
plt.title("Hazy-image")
plt.show()
```

```python
digital_img = cv2.cvtColor(hazy_img, cv2.COLOR_BGR2YCrCb)
digital_img[:, :, 0] = cv2.equalizeHist(digital_img[:, :, 0])
dehazed_img = cv2.cvtColor(digital_img, cv2.COLOR_YCrCb2BGR)

plt.imshow(dehazed_img)
plt.title("Dehazed-image")
plt.show()
```

```python
print("PSNR of Original and Hazy images:", cv2.PSNR(original_img, hazy_img))
print("PSNR of Original and DeHazed images:", cv2.PSNR(original_img, dehazed_img))
```
